```python
from icrawler.builtin import GoogleImageCrawler

path='./korean dosa mastiff'
keyword = 'korean mastiff'

filters = {'type':'photo', 'size':'medium'}

google_crawler = GoogleImageCrawler(parser_threads=2, downloader_threads=4, storage={'root_dir': path})
google_crawler.crawl(keyword=keyword, offset=0, max_num=300, min_size=(224, 224), max_size=(None), filters=filters)
```
